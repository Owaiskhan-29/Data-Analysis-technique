<h3><font color="meganta">set_index()</font></h3>


>We know the index column is automatically generated. If we do not want this column and we want to set a column from our data as index column, that is possible using set_indexi method. The column with unique values can be set as index column. For example, to make 'MovieName' column as index column for our data frame, we can write:

```python
df.set_index("MovieName")
```

>We can find the MovieName being used as index column in the new data frame 'df' as above. However, the original data frame 'df in this case is not modified and it still uses automatically generated index column,below we can observe it...

>If we want to modify the original df and set MovieName as index column, we should add " inplace=True " as shown below:

```python
df.set_index("MovieName",inplace=True)
```
