```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

# hist()

```python
plt.hist(df["tip"],color="teal",bins=10)
```

# distplot()

```python
sns.distplot(df.tip,hist=True,kde=True)
```

# displot()

```python
sns.displot(data=df,x="tip",kde=True,hue="sex",col="day")
```
