## Data Cleaning 

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df['Order_Date'] = pd.to_datetime(df['order_date'])
```

```python
df['Order_Date'].astype
```

```python
df1 = df.drop('order_date', axis=1)
```

```python
# Recalculate 'placed_gmv' based on the formula: final_ak_price * units_sold * quantity
df1['calculated_gmv'] = df1['final_ak_price'] * df1['units_sold'] * df1['quantity']
```

```python
# Check if any discrepancies exist between 'placed_gmv' and 'calculated_gmv'
discrepancies = df1[df1['placed_gmv'] != df1['calculated_gmv']]
```

```python
df1
```

```python
 #Display discrepancies if any
discrepancies.head(), discrepancies.shape
```

```python
# Correct the 'placed_gmv' column with the recalculated values
df1['placed_gmv'] = df1['calculated_gmv']
```

```python
# Drop the 'calculated_gmv' column
df1.drop(columns=['calculated_gmv'], inplace=True)
```
