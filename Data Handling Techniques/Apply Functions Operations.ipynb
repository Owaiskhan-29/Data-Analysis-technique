```python
def count_missing(vec):
    isnull_ = pd.isnull(vec)
    
    nan_count = np.sum(isnull_)
    
    return nan_count   
    
```

```python
df.apply(count_missing, axis = 0)
```

```python
df['num_missing'] = df.apply(count_missing, axis = 1)
```

```python
# Proportionate of missing values

def prop_missing_values(vec):
    num = count_missing(vec)
    
    dem = vec.size
    
    return num/dem;
```

```python
df.apply(prop_missing_values, axis = 0)
```

```python
df.apply(prop_missing_values, axis = 1)
```

```python
# proportionate of true value

def prop_true_values(vec):
    return 1 -prop_missing_values(vec)
```

```python
df.apply(prop_true_values)
```

```python
import re

p = re.compile('\w+\s+\w+')
```

```python
def get_names(vec):
    return p.match(vec).group();
```

```python
df['names'] = df['William Hartnell (1963-66)'].apply(get_names)
```
