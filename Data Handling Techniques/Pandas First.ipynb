### $Introduction$
    
   
    
    Pandas ia an open source Python Library for data analysis. It give Python the ability to work with spread-sheet like data for fast data loading, manipulating, transformating among other functions... To give Python these enhanced features, Pandas has introduced two NEW DATA TYPES:
    1. DataFrame (When you refer to complete data or more than one column data is called as DataFrame)
    2. Series (When you refer specific column or row is called as Series)
