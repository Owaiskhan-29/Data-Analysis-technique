```python
df = pd.read_csv("https://raw.githubusercontent.com/harissriramoju/DataScience-Samples/master/banklist.csv", parse_dates = [5,6])
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# Obtain the quarter and year in which the bank is closed
df['closing_quarter'], df['closing_year'] = df['Closing Date'].dt.quarter, df['Closing Date'].dt.year
```

```python
df.head()
```

```python
df.tail()
```

```python
# How many banks were closed in each year

closing_year = df.groupby('closing_year').size()
```

```python
# Calculate how many banks closed in each quarter of each year

closing_year_q = df.groupby(['closing_year', 'closing_quarter']).size()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax = closing_year_q.plot()
```

```python
fig, ax = plt.subplots()
ax = closing_year.plot()
```
