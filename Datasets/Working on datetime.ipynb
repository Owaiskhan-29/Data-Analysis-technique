>>> # Working on Datetime 

```python
import pandas as pd
```

```python
df=pd.read_csv("united states_excess_deaths.csv")
```

```python
df
```

```python
df=df[["region","start_date"]]
```

```python
df.dtypes
```

here the date column is the form of object type time convert into date time

```python
df["start_date"]=pd.to_datetime(df["start_date"])
```

```python
df.dtypes
```

there we can see that we successfully converted object type data into date time format

> # extraxting the month name

```python
df["month_name"]= df["start_date"].dt.month_name()
```

```python
df
```

we can see that we successfullly extracted the months name as a separate column from the date column

> # extraxting the month number

```python
df["month_number"]=df["start_date"].dt.month
df
```

> # extraxting the day name

```python
df["day_name"]=df["start_date"].dt.day_name()
df
```

> # extraxting the day number

```python
df["day_no"]=df["start_date"].dt.day
```

```python
df
```

> # extracting the week number

```python
df["day_of_week"]=pd.DatetimeIndex(df["start_date"]).day_of_week
```

```python
df
```

> # extracting the year

```python
df["year"]=pd.DatetimeIndex(df["start_date"]).year
```

```python
df
```
